```python
import ipdb
from nltk import data
from nltk.corpus.reader import CHILDESCorpusReader
from nltk.probability import FreqDist
from numpy import array, log2, sum
from pandas.tools.plotting import scatter_matrix
import pandas as pd
corpus_root = data.find('corpora/childes/English-UK-MOR/')
eng = CHILDESCorpusReader(corpus_root, 'Manchester/.*.xml')
nmmfile = "NLTKCHILDES/morph-eng.res"
nmsfile = "NLTKCHILDES/syntax-eng.res"
import matplotlib.pyplot as plt
%matplotlib inline
files = eng.fileids()
# ipdb.set_trace()
print "hello"
```

```python
resDict = {}
for f in files:
    age = eng.age(f,month=True)[0]
    child = f.split("/")[1]
    if resDict.has_key((child,age)):
        resDict[child,age] += [f]
    else:
        resDict[child,age] = [f]
print resDict
```

```python
# Computes the entropy of a FreqDist
def Entropy(d,method="ML"):
    v = array([float(elem) for elem in d.values()])
    # Corpus size
    N = sum(v)
    if len(v)>1:
        # Word probabilities
        p = v/N
        if method == "ML": 
        # Maximum-likehood method (i.e., unsmoothed entropy)
            return -sum(p*log2(p))
        # Chao-Shen method (i.e., Good-Turing smoothed, coverage-adjusted)
        else:# elif method == "Chao-Shen"
            # Number of hapax legomena
            f1 = float(len(d.hapaxes()))
            # Coverage factor
            C = 1.0 - f1/N
            if C>0: #(can only be done if the coverage is positive, otherwise probs become zero)
                # Correct probabilities (Good-Turing)
                p = C*p
            # Coverage-adjusted entropy, notice the denominator correction
        return -sum( p * log2(p) / (1.0 - (1.0-p)**N))
    
    else: # If there’s less than 2 distinct words, there is no entropy in any case
        return 0.0
```

```python
result_lexical = {}
for child,age in resDict.keys():
    result_lexical[child] = []

for child,age in resDict.keys():
    ff = resDict[child,age]
    # Frequency dists of child and mother for this particular child and age (raw)
    fchild = FreqDist(eng.words(ff,speaker="CHI",replace=True))
    fmother = FreqDist(eng.words(ff,speaker="MOT",replace=True))
    # Frequency dists of child and mother for this particular child and age (stemmed)
    # Notice that we split words on "-" to transform "be-3S" to "be"
    fchildS = FreqDist([w.split("-")[0] for w in eng.words(ff,speaker="CHI",replace=True,stem=True)])
    fmotherS = FreqDist([w.split("-")[0] for w in eng.words(ff,speaker="MOT",replace=True,stem=True)])
    # Statistics
    nchild = sum(array(fchild.values()))
    nmother = sum(array(fmother.values()))
    # Entropies
    Hchild = Entropy(fchild,method="Chao-Shen")
    Hmother = Entropy(fmother,method="Chao-Shen")
    # Entropies (stemmed)
    HchildS = Entropy(fchildS,method="Chao-Shen")
    HmotherS = Entropy(fmotherS,method="Chao-Shen")
    print child,age,nchild,Hchild,HchildS,nmother,Hmother,HmotherS
    result_lexical[child].append({'age': age,
                                  'nchild': nchild,
                                  'Hchild': Hchild,
                                  'Hchilds':HchildS,
                                  'nmother':nmother,
                                  'Hmother':Hmother,
                                  'Hmothers':HmotherS})
        
```

```python
averages = {}
fig = plt.figure(figsize=(10,10))
for child in result_lexical.keys():
    X = []
    Y = []
    for session in result_lexical[child]:
        X.append(session['age'])
        Y.append(session['Hmothers'])
        if not averages.get(session['age']):
            averages[session['age']] = session['Hmothers']
        else:
            averages[session['age']] += session['Hmothers']    
    X.sort()
    plt.plot(X, Y, linestyle='-')
    plt.title(child)
    plt.axis([20,38.0,4.0,10.0,])
# for k,v in averages.iteritems():
#     plt.plot(k,v)
```
